#  Predict Wind Power using Data
- Data has 30 months of Turbine Data recorded at an interval of 10 mins


```python
# import libraries
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
# Reading Data set and transforming it to DataFrame
Turbine_Data=pd.read_csv('Turbine_Data.csv')
Turbine_Data_df=pd.DataFrame(Turbine_Data)
Turbine_Data_df
```

# Drop NaN values


```python
Turbine_Data_df=Turbine_Data_df.dropna() #emove NaN values
Turbine_Data_df.drop_duplicates(inplace=True)
Turbine_Data_df
```

```python
Turbine_Data_df.rename(columns={'Unnamed: 0':'Time-stamp'},inplace=True)
Turbine_Data_df
```

```python
#check for missing values
Turbine_Data_df.isna().sum()
```

```python
Turbine_Data_df.info()
```

# converting time-stamp to datetime

```python
for val in Turbine_Data_df['Time-stamp']:
    date_obj=pd.Timestamp(val)
    date=date_obj.date() 
    minutes=date_obj.minute
    Turbine_Data_df['Date']=date  
    Turbine_Data_df['minutes']=minutes
```

```python
Turbine_Data_df.drop(columns=['Time-stamp'],inplace=True)
Turbine_Data_df
```

```python
Turbine_Data_df.index.name='S.no'
Turbine_Data_df.index=Turbine_Data_df['ActivePower']
```

```python
Turbine_Data_df
```

```python
 Turbine_Data_df['year']=[val.year for val in Turbine_Data_df['Date']]
 Turbine_Data_df['month']=[val.month for val in Turbine_Data_df['Date']]
 Turbine_Data_df['day']=[val.day for val in Turbine_Data_df['Date']]
 Turbine_Data_df.index=Turbine_Data_df['year']
```

```python
Turbine_Data_df.drop(columns=['year'],inplace=True)
```
